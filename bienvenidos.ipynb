# Bienvenidos!

**Bienvenidos en el OGGM-Edu JupyterLab**! Aqu se puede realizar experimentos con "glaciares digitales": sin hielo pero interactivos y menos peligrosos que los glaciares Peruanos.

Es probable que ese entorno "JupyterLab" es nuevo para ti, pero no te preocupes! Aqu nunca te puedes equivocar: siempre se puede empezar de nuevo.

En esta pagina te enseamos las caractersticas fundamentales del entorno "JupyterLab" de OGGM-Edu: para mas explicaciones, escucha las recomendaciones de tu profesora.

## Estructura

El JupyterLab tiene tres zonas principales: la **barra del men** (azul), el **explorador de archivos** (rojo), y las **pestaas con los cuadernos activos** (verde):

## Abrir, cerrar y guardar los cuadernos

En el **explorador de archivos** se puede seleccionar los cuadernos: haz un "doble clic" en el cuaderno que quieres abrir.

Para cerrar un cuaderno, usa la cruz de la pestaa (como en tu navegador web). Si haz cambiado algo en tu cuaderno, veras que la cruz esta escondida por un circulo gris. Para guardar tus cambios, usa el botn como enseado en la video abajo, o usa `CTRL+S`.

## Copiar un archivo

Copiar los cuadernos antes de cambiarles puede ser muy til: por ejemplo para hacer experimentos sin perder los archivos originales.

**Por eso te recomendamos de renombrar los cuadernos antes de hacer los experimentos**. As, siempre puedes volver a ver al cuaderno original. Copiar un cuaderno es muy fcil: haz un clic derecho en el archivo y selecciona "Duplicate". Despues, renombra el archivo como te lo ensea esta video: 

*Duplicate/rename/edit/delete video*:

## Volver al estado original

Si por algn razn quieres volver al estado original del curso (por ejemplo si has perdido los cuadernos originales, o si la profesora te recomend de hacerlo), es muy simple:

1. Borra o renombra el directorio de trabajo
2. Cerra tu entorno "JupyterLab"
3. brelo otra vez desde la pagina web CdeC-glaciologia: https://github.com/ehultee/CdeC-glaciologia
